**AIM :** Implementing Support Vector Classification to check whether the person gets the promotion based on the attributes Age and ExpectedSalary.

"Importing the required modules."

```python
from sklearn.svm import SVC
import pandas as pd
import matplotlib.pyplot as plt
```

"Loading the dataset into our environment."

```python
dataset=pd.read_csv('/content/logistic regression dataset-Social_Network_Ads.csv')
```

"Checking whether the dataset have any values rather than numericals."

```python
dataset.info()
```

"Separating the input and output attributes from the dataset."

```python
X=dataset[['Age','EstimatedSalary']]
```

```python
y=dataset[['Purchased']]
```

```python
X
```

```python
y
```

"Importing the selection module from sklearn."

```python
from sklearn.model_selection import train_test_split
```

"Splitting the dataset for the purpose of testing and training."

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=40)
```

"Calling the SVC algorithm as a constructor."

```python
clf=SVC(kernel='rbf')
```

"Fitting the data to the model."

```python
clf.fit(X_train,y_train)
```

"Get support vectors."

```python
clf.support_vectors_
```

"Get indices of support vectors."

```python
clf.support_
```

"Get the number of support vectors for each class"

```python
clf.n_support_
```

"Storing the predicted values of training data."

```python
y_pred=clf.predict(X_train)
```

```python
y_pred
```

"Importing all the required model evaluating metrics."

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

"Calculating the accuracy, confusion matrix and classification report."

```python
acc=accuracy_score(y_train,y_pred)
print(acc)
matrix=confusion_matrix(y_train,y_pred)
print(matrix)
report=classification_report(y_train,y_pred)
print(report)
```

"Storing the predicted values of testing data."

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

"Calculating the accuracy, confusion matrix and classification report."

```python
acc=accuracy_score(y_test,y_pred)
print(acc)
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
report=classification_report(y_test,y_pred)
print(report)
```
